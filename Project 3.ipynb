```python
# Import packages
import pandas as pd
import seaborn as sns
import numpy as np
import scipy.stats as stats
```

### EDA


```python
# Import dataset
states = pd.read_excel('Desktop/Comp bio/USArrests w state names.xlsx')
```

```python
# peak at dataset
print(states.head())
# get info on dataset
print(states.shape)
```

*There are 50 rows and 10 columns. One row represents an observation of a state with their respecitve values for the variables and what region they are in.*

```python
# describe states
states.describe()
```

*The mean amount of arrests for murder in the US was 7.78 and the mean amount of total arrests in the US was 199.78. The state with the most arrests had 382.3.*

```python
# Find mean and std of numerical varaibles  by region
(states.groupby(['Region'])
.agg(['mean','std','size']))
```

*The South region had the largest mean for murder and assault arrests. The West region had the largest mean for rape arrests.* 

### Visualizations

```python
sns.pairplot(states, hue='Region')
```

Based on the many visuals, there is one strong correlation present. There is a strong linear reltionship between Total arrests and Assualt arrests. This makes sense due to assualts being the majority of crimes.

```python
sns.barplot(data=states, x='Region', y='Total')
```

As evident from the bar chart above, the South and West region have way more total arrests than the Northeast and North Central regions.
